```python
def circfun(xc, yc, radius, npoint):
    theta = np.linspace(-np.pi, np.pi, npoint)
    x = radius*np.cos(theta)
    y = radius*np.sin(theta)
    return x+xc, y+yc
```

```python
def gaussian(x,a,b,c,d):
    return a*np.exp(-(x-b)**2/(2*c**2))+d
```

```python
xanti = np.arange(41)*1
yanti = gaussian(xanti, 8., 20., 3., 7.5)
```

```python
fig, ax = plt.subplots(1,1, figsize = (10, 6))
ax.plot(xrecv, yrecv, 'ko', ms =10)
ax.plot(xanti, -yanti, 'k--', lw=2)
ax.plot(np.r_[0., 40.], np.r_[0., 0.], 'k-', lw=2)
ax.set_yticks(-np.arange(0,42,2.5)+2.5)
ax.tick_params(axis="y", pad=4)
ax.set_xticks(np.arange(0,41,4))
ax.set_ylabel('Depth (m)', fontsize = 18)
ax.xaxis.tick_top()
ax.set_xlabel('x (m)', fontsize = 18)
ax.xaxis.set_label_coords(0.5,1.12)
ax.set_ylim(-20., 2.5)
ax.grid(True)
fig.savefig('anticline.png')
```
